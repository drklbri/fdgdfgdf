### Импорт библиотек
Важной особенностью Python как экосистемы, по сравнению с просто языком программирования, является наличие большого количества библиотек и инструментов. Эти библиотеки и инструменты обычно должны быть импортированы при необходимости (например, библиотека построения графиков) или запущены как отдельный системный процесс (например, среда разработки Python). Те из вас, кто знаком с  Matlab и R, в курсе, что они поставляются с предварительно упакованным набором модулей различных аспектов научных вычислений. Импорт означает предоставление библиотеки текущему пространству имен и текущему процессу интерпретатора Python. Используется зарезервированное слово import; так же, как и в C++, где префикс #include является синтаксисом для импорта библиотеки.

Сам Python уже поставляется с большим набором библиотек, которые расширяют базовый интерпретатор в разных направлениях. Например, основные математические вычисления; ввод-вывод; манипуляции со строками могут быть выполнены без какого-либо импорта, в то время как операции, требующие более сложных математических функций, должны быть импортированы через математическую библиотеку. Когда программа Python запускается, она имеет доступ только к базовым функциям и классам.
“Модули" содержат дополнительную функциональность. Используйте команду "import”, для загрузки модуля

## Операции со строками

Две строки могут быть соединены/сложены вместе (объединены) с помощью оператора * * + * * :

```python
"Hello, "+"world!"
```

Это соединение строк очень просто; однако, если вы хотите, чтобы слова разделялись пробелом, вы должны поместить этот пробел. Вот несколько примеров

```python
"Hello, "  +  "world!" # пробел после запятой
```

```python
"Hello,"  +  " world!" # пробел перед вторым словом
```

```python
"Hello,"  +  "world!" # без пробелов
```

### Boolean арифметика

Boolean тип имеет свою собственную арифметику, как и обычные числа. Числа имеют арифметические операции +, –, ××, ÷÷. Какие операции выполняют boolean операторы? Первая операция - это оператор **and**.

True **and** True $\longrightarrow$ True 

True **and** False $\longrightarrow$ False 

False **and** True $\longrightarrow$ False 

False **and** False $\longrightarrow$ False

**and** из двух booleans значений True если (и только если) оба изначально True. Если оба False тогда результат False. Мы видим что **and** условие сильное.

**Примеры**

```python
5 < 10 and 6 < 8
```

Давайте посмотрим, что здесь происходит. 5 < 10 True **and** 6 < 8 true , следовательно, результат true.

```python
5!=10 and 6 > 8
```

Похожее нарушение дает 5$\neq$10 $\longrightarrow$ True **and** 6 > 8 $\longrightarrow$ False, следовательно результат False.

Теперь посмотрим на оператор **or** 

True **or** True $\longrightarrow$ True 

True **or** False $\longrightarrow$ True 

False **or** True $\longrightarrow$ True 

False **or** False $\longrightarrow$ False

Результат этой операции True если оба начальные значения True и False если только одно False. Условие **or** слабее чем **and**.

**Примеры**

```python
5 < 10 or 6 < 8 # это сводится к 'True or True'
```

```python
5!=10 or 6 > 8 # это сводится к 'True or False'
```

```python
5==10 or 6 > 8 # это сводится к 'False or False'
```

Финальный оператор - **not**. Он принимает только одно начальное значение и “зеркалит” его. True становится False и наоборот.

**not** True $\longrightarrow$ False 

**not** False $\longrightarrow$ True 

**Примеры**


```python
not 6>7 # оценивает, чтобы не было ложно
```

```python
not 5!=10 # равно не правда
```

## Цикл While
Ключевое слово while позволяет повторять операцию, пока выполняется условие.

Ниже код

```python
number = 1
while(number <= 10):
    print(number)
    number += 1
print("Done") # необъодимо оставить цикл для завершения вложения отступа 
    
```

Линия 1: старт цикла с начальным значением

Линия 2: зарезервированное слово **while** в месте, которое говорит Python, что есть цикл, за которым нужно следовать. За ним должно непосредственно следовать выражение, которое вычисляется как логическое; т. е. номер условия теста $\leq$10, которое является выражением Python, которое вычисляется как логическое - истинное или ложное. За условием теста следует: этот отступ для всего последующего кода в блоке с 4 пробелами.

Предположим, мы хотели бы знать, в течение скольких лет мы должны хранить 100 фунтов на сберегательном счете, чтобы достичь 200 фунтов просто за счет ежегодной выплаты процентов по ставке в 5%. Вот программа, позволяющая рассчитать, что на это уйдет 15 лет:

```python
mymoney = 100 # в GBP
rate = 1.05 # 5% ставка
years = 0
while mymoney < 200: # повторить до тех пор пока не будет достигунто 200 
    mymoney = mymoney * rate
    years += 1
print('Нам нужно ', years , 'лет для достижения ', int(mymoney) , 'фунтов .')
```

### Оператор with 
Это тот самый пример , который не требует явного закрытия кода python и считается более безопасным.

```python
with open('/Users/User/Desktop/Humpty.txt') as f: 
    rhyme = f.read()
print(rhyme)
```

```python
with open('/Users/User/Desktop/Humpty.txt') as f:
    rhyme = f.readline()
    print(rhyme) # выводит первую линию
    rhyme = f.readline()
    print(rhyme) # выводит линию номер 2
```

Следующий скрипт показывает эффективный способ чтения в текстовом файле строка за строкой

```python
with open('/Users/User/Desktop/Humpty.txt','r') as f:
    for line in f:
        print(line)
```

Приведенный выше код эффективен, потому что все содержимое файла не читается за один раз. Вместо этого он повторяет текст, захватывая одну строку за другой. Очевидно, что есть способы дать пользователю больший контроль над тем, что он читает.

#### Функция расчета факториала

```python
def factorial(n):
    factorial = 1 # базовый случай
    for n in range(1,n+1): # помните, что для цикла требуется отступ
            factorial = factorial*n
    return factorial

n=int(input("введите значение integer: "))

print("факториал ", n, "= ",factorial(n))
```

```python
def factorial(n):
    num = 1
    while n >= 1:
        num = num * n
        n = n - 1
    return num

answer = factorial(0)
print(answer)
```

### Приложения для исчислений

## Нахождение корней

```python
from scipy import optimize # функция нахождения корней в этом модуле
import numpy as np

def function(x):
    return (x-exp(1/x))

# N-R: найти 0 f(x) с начальным предположением x=1
value1 = optimize.newton(function,1)
print("N-R gives the root as ", value1)

# найти 0 f(x) между (1,2) путем деления пополам
value2 = optimize.bisect(function,1,2)
print("Bisection gives ", value2)
```

### Получаем $N(0,1)$ из $U(0,1)$
Мы знаем взаимотношения между $N(x)$ и $erf(x)$ 
\begin{equation*}
N\left( x\right) =\frac{1}{2}\left( 1+{erf}\left( \frac{x}{\sqrt{2}}
\right) \right) 
\end{equation*}
Переопределяя это дает

$$
x=\sqrt{2}{erf}^{-1}( 2y-1) ;\ y\sim U\left( 0,1\right) 
$$

где $x\sim N\left( 0,1\right).$



```python
import numpy.random as npr
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
x = np.zeros(100000)

from scipy.special import erfinv
for i in range(100000):
    x[i] = sqrt(2)*erfinv(2*(npr.uniform(0,1))-1)
plt.hist(x,bins = 100, cumulative = False, color = 'red')
plt.show()
```
